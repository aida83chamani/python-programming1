```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
x1 = np.arange(0, 1, 0.1)
x2 = np.arange(0, 1, 0.01)
y1 = np.sin(2 * m.pi * x2)
y2 = [0.125, 0.5, 0.95, 0.92, 0.625, -0.125, -0.7, -0.92, -0.98, -0.68]
plt.plot(x2, y1, label='true underlying function')
plt.scatter(x1, y2, color='blue', label='observed data')
plt.axis([0, 1, -1, 1])
plt.title('Plot of noise corrupted data')
plt.legend()
plt.show()
```

```python
def is_monotonic(a):
    increasing = True
    decreasing = True

    for i in range(len(a) - 1):
        if a[i] < a[i + 1]:
            decreasing = False
        elif a[i] > a[i + 1]:
            increasing = False

    if increasing or decreasing:
        return 'monotonic'
    else:
        return 'not monotonic'
print(is_monotonic([1,2,3,3,3,3,3,4,5,6,7]))
print(is_monotonic([7,6,5,4,3,3,3,3,2,1]))
print(is_monotonic([1,5,4,3,3,3,3,3,2,6,7]))
```
